This section of the code imports the dataset from the website and sorts the data by year. It also prints the resulting dataset.

```python
import pandas as pd
from matplotlib import pyplot as plt

#Importing dataset from website
Health_Survey = pd.read_csv('https://raw.githubusercontent.com/Bowen-cancodefinialproject/PFDAFINIALPROJECT/main/New_York_City_Community_Health_Survey.csv').sort_values("Year")

#Printing data
Health_Survey
```

The following section finds the mean, median and mode for each column of sata in the dataset and prints it in tabular form.

```python
#Finding the mean, median and mode for each column of data
a=[Health_Survey['No Health Insurance'].mean(),Health_Survey['No Health Insurance'].median(), Health_Survey['No Health Insurance'].mode().tolist()]
b=[Health_Survey['Did not get needed medical care'].mean(),Health_Survey['Did not get needed medical care'].median(),Health_Survey['Did not get needed medical care'].mode().tolist()]
c=[Health_Survey['No Personal Doctor'].mean(),Health_Survey['No Personal Doctor'].median(),Health_Survey['No Personal Doctor'].mode().tolist()]
d=[Health_Survey['Drinks 1 or more sugar-sweetened beverages per day'].mean(),Health_Survey['Drinks 1 or more sugar-sweetened beverages per day'].median(),Health_Survey['Drinks 1 or more sugar-sweetened beverages per day'].mode().tolist()]
x=[Health_Survey['Smoking Status (current smokers)'].mean(),Health_Survey['Smoking Status (current smokers)'].median(),Health_Survey['Smoking Status (current smokers)'].mode().tolist()]
f=[Health_Survey['Binge Drinking'].mean(),Health_Survey['Binge Drinking'].median(),Health_Survey['Binge Drinking'].mode().tolist()]
g=[Health_Survey['Obesity'].mean(),Health_Survey['Obesity'].median(),Health_Survey['Obesity'].mode().tolist()]
h=[Health_Survey['Colon cancer screening, adults age 50+ (colonoscopy)'].mean(),Health_Survey['Colon cancer screening, adults age 50+ (colonoscopy)'].median(),Health_Survey['Colon cancer screening, adults age 50+ (colonoscopy)'].mode().tolist()]
i=[Health_Survey['Self-reported Health Status (excellent/very good/good)'].mean(),Health_Survey['Self-reported Health Status (excellent/very good/good)'].median(),Health_Survey['Self-reported Health Status (excellent/very good/good)'].mode().tolist()]
j=[Health_Survey['Flu shot in last 12 months, adults ages 65+ (not age-adjusted)'].mean(),Health_Survey['Flu shot in last 12 months, adults ages 65+ (not age-adjusted)'].median(),Health_Survey['Flu shot in last 12 months, adults ages 65+ (not age-adjusted)'].mode().tolist()]

#Assigning the corresponding data calculated above with the correct title to place in tabular form
data= {'No Health Insurance': a,
        'Did not get needed medical care':b,
        'No Personal Doctor':c,
       'Drinks 1 or more sugar-sweetened beverages per day':d,
       'Smoking Status (current smokers)':x,
       'Binge Drinking':f,
       'Obesity':g,
       'Colon cancer screening, adults age 50+ (colonoscopy)':h,
       'Self-reported Health Status (excellent/very good/good)':i,
       'Flu shot in last 12 months, adults ages 65+ (not age-adjusted)':j
}

#Placing data calculated in tabular form and labelling the rows for the table
df = pd.DataFrame(data,index=['mean','median','mode'])

#Printing out calculated mean, median and mode in a table
df

```

This section of code identifies and removes duplicate rows from the dataset

```python
# Finding how many duplicate rows exist in dataset
print('There are', Health_Survey.duplicated().sum(), 'duplicate rows.')

#Removing any duplicate rows in the existing dataset
Health_Survey.drop_duplicates(inplace=True)
```

In this section a scattered plot is graphed to show how binge drinking in connected to obesity from this daa collected. Although there are some outliers this plot shows a linear relationship between binge drinking and obesity between 2010 and 2021.

```python

import matplotlib.pyplot as plt

#Assigning data for x and y axis
x = Health_Survey['Binge Drinking']
y = Health_Survey['Obesity']

#Adding a title for the plot and labelling the axis
plt.title('Scatter chart')
plt.xlabel('Binge Drinking')
plt.ylabel('Obesity')

#plotting the graph
plt.scatter(x, y )
plt.show()
```

In the following section an analysis is done to show how binge drinking by New Yorkers changed over the span 2010-2021 by using a line graph. From 2010-2014 binge drinking linearly increased and then decreased, from then there was a logarithmic growth showing potential of eventually plateaus.

```python
from traitlets.traitlets import DottedObjectName
import matplotlib.pyplot as plt
import numpy as np

#Extracting Columns of data to be used for analysis
ypoints = Health_Survey.filter( ["Year", "Binge Drinking"])

#Filtering rows od data to be used.
y= pd.DataFrame(ypoints.filter(items=[27,24,21,18,15,12,9,6,3,0,30],axis=0))

#Adding title and labeling axes
plt.title('Binge Drinking from 2010-2021')

plt.ylabel('Binge Drinking')
plt.xlabel('Year')

#plotting line graph to show how Binge drinking changed from 2010-2021
#while also changng the linestyle

plt.plot(y['Year'],y['Binge Drinking'] ,linestyle = 'dotted')

plt.show()
```
